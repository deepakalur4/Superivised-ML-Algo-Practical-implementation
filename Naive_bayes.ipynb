```python
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset=load_iris()
```

```python
dataset
```

```python
dependent_features=pd.DataFrame(dataset["data"],columns=dataset.feature_names)
independent_features=pd.DataFrame(dataset["target"],columns=["Flower_name"])
```

```python
dependent_features.shape
```

```python
dependent_features
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
scaling=StandardScaler()
X_train,X_test,y_train,y_test=train_test_split(dependent_features,independent_features,test_size=0.2,random_state=42)
```

```python
X_train=scaling.fit_transform(X_train)
X_test=scaling.transform(X_test)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
```

```python
model=GaussianNB()
model_2=KNeighborsClassifier(n_neighbors=3)
model_3=SVC()
```

```python
model.fit(X_train,y_train)
```

```python
preds=model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
import seaborn as sns
```

```python
print(accuracy_score(preds,y_test))
```

```python
sns.heatmap(confusion_matrix(preds,y_test),cmap="viridis",fmt="d",annot=True)
```

```python
model_2.fit(X_train,y_train)
```

```python
model_2.predict(X_test)
```

```python
preds=model_2.predict(X_test)
```

```python
print(accuracy_score(preds,y_test))
```

```python
sns.heatmap(confusion_matrix(preds,y_test),cmap="viridis",fmt="d",annot=True)
```

```python
print(classification_report(preds,y_test))
```

```python
model_3.fit(X_train,y_train)
```

```python
model_3.predict(X_test)
```

```python
preds=model_3.predict(X_test)
```

```python
print(accuracy_score(preds,y_test))
```
